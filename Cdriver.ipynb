```python
xudoc=pq(xu)
xulst=xudoc('#content_left')
#xlst=xulst('.result-op c-container xpath-log')
for i in range(1,11):
    s='#'+str(i)+' h3'
    #print(s)
    xlst=xulst(s).text()#获得搜索结果标题
    print(xlst)
```

```python
x2=xudoc('#page a').text()
xl=xudoc('#page a')
print(x2)
for i in range(0,9):
    print(pq(xl[i]).attr('href'))#打印跳转页面的url
```
